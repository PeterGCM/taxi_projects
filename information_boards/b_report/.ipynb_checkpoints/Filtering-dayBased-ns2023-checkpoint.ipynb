```python
import __init__
#
from IPython.display import HTML, display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_dist(df, cn, m, sd):
    print 'mean %.3f' % m
    print 'sigma %.3f; 2sigma %.3f; 3sigma %.3f;' % (sd, 2 * sd, 3 * sd)
    df[cn].hist()
```

```python
from information_boards import statisticsAllDrivers_ns_dpath
from information_boards import statisticsAllDriversDay_ns2023_prefix
#
fnDayNS = '%sall.csv' % statisticsAllDriversDay_ns2023_prefix
dfDayNS = pd.read_csv('%s/%s' % (statisticsAllDrivers_ns_dpath, fnDayNS))
```

```python
print len(dfDayNS)
outlier_index = set()
for cn in dfDayNS.columns:
    if cn in ['year', 'month', 'day', 'driverID']:
        continue
    if cn == 'wleProductivity':
        outlier_set = set(np.where(dfDayNS[cn] > 80)[0].tolist())
        outlier_index = outlier_index.union(set(outlier_set))                                        
    if cn == 'QTime/locTrip':
        outlier_set = set(np.where(dfDayNS[cn] > 120)[0].tolist())
        outlier_index = outlier_index.union(set(outlier_set))
    outlier_set = np.where((np.abs(dfDayNS[cn] - dfDayNS[cn].mean()) > (3 * dfDayNS[cn].std())))[0]
    outlier_index = outlier_index.union(set(outlier_set))
print len(outlier_index)
dfDayNS = dfDayNS.drop(dfDayNS.index[list(outlier_index)])
```
