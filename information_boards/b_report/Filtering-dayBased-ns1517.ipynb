```python
import __init__
#
from IPython.display import HTML, display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_dist(df, cn, m, sd):
    print 'mean %.3f' % m
    print 'sigma %.3f; 2sigma %.3f; 3sigma %.3f;' % (sd, 2 * sd, 3 * sd)
    df[cn].hist()
```

```python
from information_boards import statisticsAllDrivers_ns_dpath
from information_boards import statisticsAllDriversDay_ns1517_prefix
#
fnDayZ = '%sall.csv' % statisticsAllDriversDay_ns1517_prefix
dfDayZ = pd.read_csv('%s/%s' % (statisticsAllDrivers_ns_dpath, fnDayZ))
```

```python
print len(dfDayZ)
outlier_index = set()
for cn in dfDayZ.columns:
    if cn in ['year', 'month', 'day', 'driverID']:
        continue
    if cn == 'wleProductivity':
        outlier_set = set(np.where(dfDayZ[cn] > 80)[0].tolist())
        outlier_index = outlier_index.union(set(outlier_set))                                        
    if cn == 'QTime/locTrip':
        outlier_set = set(np.where(dfDayZ[cn] > 120)[0].tolist())
        outlier_index = outlier_index.union(set(outlier_set))
    outlier_set = np.where((np.abs(dfDayZ[cn] - dfDayZ[cn].mean()) > (3 * dfDayZ[cn].std())))[0]
    outlier_index = outlier_index.union(set(outlier_set))
print len(outlier_index)
dfDayZ = dfDayZ.drop(dfDayZ.index[list(outlier_index)])
```
