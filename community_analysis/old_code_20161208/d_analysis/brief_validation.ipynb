```python
for x in range(8):
    group_fpath = '%s/%s/%s' % (group_dpath, 'fb', 'group-fb-2009-G(%d).pkl' % x)
    groupGraph = ig.Graph.Read_Pickle(group_fpath)
    groupGraph_set = set([(groupGraph.vs[i]['name'], groupGraph.vs[j]['name']) for (i, j) in groupGraph.get_edgelist()]) 
    causalityGraph_fpath = '%s/%s/%s' % (causalityGraph_dpath, 'fb', 'causalityGraph-fb-2009-G(%d).pkl' % x)
    causalityGraph = load_pickle_file(causalityGraph_fpath)
    causalityGraph_set = set(causalityGraph)
    print 'Group %d: # of drivers %d' % (x, len(groupGraph.vs))
    print '    # of edges by frequency and benefit function:', len(groupGraph_set)
    print '    # of edges by regression:', len(causalityGraph_set)
    print '    Percentage %.3f' % (len(groupGraph_set.intersection(causalityGraph_set)) / float(len(groupGraph_set)))
```
