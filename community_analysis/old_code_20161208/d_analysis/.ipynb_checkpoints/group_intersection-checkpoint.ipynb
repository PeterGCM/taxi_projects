```python
import __init__
#
from community_analysis import group_dir
#
import igraph as ig
import numpy as np
#
month3_group = {}
for month3 in ['M01M02M03', 'M02M03M04']:
    for i in range(11):
        pkl_fn = 'group-per990-2009-%s-G(%d).pkl' % (month3, i)
        pkl_fpath = '%s/%s' % (group_dir, pkl_fn)
        igG = ig.Graph.Read_Pickle(pkl_fpath)
        month3_group[month3, i] = [v['name'] for v in igG.vs]
whole_drivers = set()
for drivers in month3_group.values():
    for did in drivers:
        whole_drivers.add(did)
```

```python
drivers_list = list(whole_drivers)
```

```python
month3G_vec = {k: [0] * len(drivers_list) for k in month3_group.iterkeys()}
```

```python
for k, drivers in month3_group.iteritems():
    for did in drivers:
        month3G_vec[k][drivers_list.index(did)] = 1
```

```python
matching = [-1] * 11
for i in range(11):
    prevRoll = month3_group['M01M02M03', i]
    maxNumIntersect = -1e400
    maxIndex = -1
    print '2009 M01M02M03'
    print 'Group(%d): #D %d' % (i, len(prevRoll))
    print '\t 2009 M02M03M04'
    for j in range(11):
        nextRoll = month3_group['M02M03M04', j]
        s = set(prevRoll).intersection(set(nextRoll))
#         dev = np.array(month3G_vec['M01M02M03', i]) - np.array(month3G_vec['M02M03M04', j])
#         print '\t', j, len(s), (dev ** 2).sum()
        print '\t Group(%d): # intersection %d' % (j, len(s)) 
        if maxNumIntersect < len(s):
            maxNumIntersect = len(s)
            maxIndex = j
    maxIndex = maxIndex if maxNumIntersect != 0 else -1
    print 'Group(%d) is %.2f similar to Group(%d)' % (i, maxNumIntersect / float(len(prevRoll)), maxIndex)
    
        
        
    matching[i]= maxIndex
    print ''
matching
```
