```python
import __init__
from community_analysis import com_trip_dir
#
from taxi_common.file_handling_functions import get_all_files
from taxi_common.sg_grid_zone import get_sg_grid_xy_points, get_sg_zones
#
import pandas as pd
import folium
```

```python
x_points, y_points = get_sg_grid_xy_points()
zones = get_sg_zones()
```

```python
xc, yc = (x_points[0] + x_points[-1]) / float(2), (y_points[0] + y_points[-1]) / float(2)
```

```python
a = zones[(3,10)].cCoor_gps
```

```python
color_map = ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#00FFFF', '#FF00FF']
map_osm = folium.Map(location=[yc, xc], zoom_start=12)
# vertical lines
for x in x_points:
    sx, sy, ex, ey = x, y_points[0], x, y_points[-1]
    map_osm.add_children(folium.PolyLine(locations=[(sy, sx), (ey, ex)], weight=0.5))
# vertical lines
for y in y_points:
    sx, sy, ex, ey = x_points[0], y, x_points[-1], y
    map_osm.add_children(folium.PolyLine(locations=[(sy, sx), (ey, ex)], weight=0.5))

folium.Marker((yc, xc),
              popup='Mt. Hood Meadows <br>\n  test',
              icon=folium.Icon(color = color_map[0])
             ).add_to(map_osm)    
    

map_osm
```
