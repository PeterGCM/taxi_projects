```python
import __init__
#
from community_analysis import dpaths, prefixs
#
from taxi_common.file_handling_functions import get_all_files
#
import pandas as pd
import numpy as np
import statsmodels.api as sm
#
sig_level = 0.10
```

```python
gt_dpath = dpaths['spendingTime', '2009', 'groupTrips']
gt_prefix = prefixs['spendingTime', '2009', 'groupTrips']
gt_fpath = '%s/%s%s.csv' % (gt_dpath, gt_prefix, 'G(22)')
df = pd.read_csv(gt_fpath)
df
```

```python
df = df[~(np.abs(df['spendingTime'] - df['spendingTime'].mean()) > (3 * df['spendingTime'].std()))]
```

```python
zizj = '29#23'
zizj_df = df[(df['zizj'] == zizj)]
y = zizj_df['spendingTime']
X = zizj_df['priorPresence']
X = sm.add_constant(X)
res = sm.OLS(y, X, missing='drop').fit()
print res.summary()
```

```python
groupZones = {}
for zizj, pp_num in df.groupby(['zizj']).sum()['priorPresence'].iteritems():
    if pp_num < 2:
        continue
    zizj_df = df[(df['zizj'] == zizj)]
    y = zizj_df['spendingTime']
    X = zizj_df['priorPresence']
    X = sm.add_constant(X)
    res = sm.OLS(y, X, missing='drop').fit()
    if res.params['priorPresence'] < 0 and res.pvalues['priorPresence'] < sig_level:
        groupZones[zizj] = res.params['priorPresence']
groupZones
```
